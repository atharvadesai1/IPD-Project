```python
import pandas as pd
from langdetect import detect

# # Step 1: Read the CSV file into a DataFrame
# df = pd.read_csv('your_dataset.csv')

# Step 2: Detect language for each tweet and create a new column 'Language'
def detect_language(x):
  try:
    if pd.notna(x):  # Adjust the minimum length as needed
        return detect(str(x))
    else:
        return 'Not Identified'
  except:
    return 'Not Identified'

df['Language'] = df['tweetText'].apply(detect_language)
df

# # Step 3: Filter out tweets with detected language (not None)
# df_filtered = df[df['Language'].notna()]

# # Display the filtered DataFrame
# print("DataFrame with detected language:")
# print(df_filtered)

# # Save the filtered DataFrame to a new CSV file
# df_filtered.to_csv('filtered_dataset.csv', index=False)
```
